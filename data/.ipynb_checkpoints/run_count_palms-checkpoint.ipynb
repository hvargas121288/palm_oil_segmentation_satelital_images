```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.pyplot import imread

np.set_printoptions(precision = 2)
np.set_printoptions(suppress  = True)
```

```python
def normalize(img):
    img_copy  = img.copy()
    img_copy  = img_copy.astype('float64') 
    img_min   = np.min(img_copy)
    img_max   = np.max(img_copy)
    img_copy  = (img_copy - img_min)/(img_max - img_min)
    return img_copy

def convert_8bits(img):
    img_copy  = img.copy()
    img_copy  = img_copy.astype('float64') 
    img_copy  = normalize(img_copy)
    img_copy  = np.round(255*img_copy).astype('uint8')
    return img_copy
```

```python
img = imread('multi_recorte_2.png')[:,:,0]
img_norm = normalize(img)

# Convert the image to grayscale
img_norm_gauss = cv2.GaussianBlur(img_norm, (15, 15), 0)
img_norm_gauss = convert_8bits(img_norm_gauss)
```

```python
fig, axs = plt.subplots(figsize = (10,10))
axs.imshow(img_norm_gauss)
axs.axis('off')
plt.show()
```

```python
# Use the HoughCircles method to detect circles
circles = cv2.HoughCircles(
    img_norm_gauss,
    cv2.HOUGH_GRADIENT,
    dp=1.2,
    minDist=20,
    param1=50,
    param2=30,
    minRadius=10,
    maxRadius=100
)

output = np.zeros(img_norm_gauss.shape, dtype = 'float32')

# Ensure at least some circles were found
if circles is not None:
    # Convert the circle parameters (x, y, radius) to integers
    circles = np.round(circles[0, :]).astype("int")

    # Loop over the detected circles
    for (x, y, r) in circles:
        # Draw the circle in the output image
        cv2.circle(output, (x, y), r, (255, 255, 255), 2)


    # Print the number of circles detected
    print(f"Number of circles detected: {len(circles)}")
```

```python
A     = output
B     = img_norm_gauss
img_all  = np.dstack((A,B,A))

fig, axs = plt.subplots(figsize = (15,15))
axs.imshow(img_all/np.max(img_all))
axs.axis('off')
plt.show()
```
